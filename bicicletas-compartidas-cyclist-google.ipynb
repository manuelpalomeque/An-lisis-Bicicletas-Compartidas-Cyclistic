## G) verificar  las clasificaciones de los tipos de biciletas y de clientes

## H) Remplazar los valores:
Reemplace los valores al español de las columnas  tipo_bicicleta y tipo_cliente

```r
tabla_general_5 <- tabla_general_4 %>%
  mutate(tipo_bicicleta = if_else(tipo_bicicleta == "classic_bike", "bicicleta clásica", tipo_bicicleta)) %>%
  mutate(tipo_bicicleta = if_else(tipo_bicicleta == "electric_bike", "bicicleta eléctrica", tipo_bicicleta)) %>%
  mutate(tipo_bicicleta = if_else(tipo_bicicleta == "docked_bike", "bicicleta docked", tipo_bicicleta))  %>%
  mutate(tipo_cliente = ifelse(tipo_cliente == "member", "miembro", tipo_cliente))
```

## I) Filtrar datos:
Como el analisis busca centrarse en los miembros casuales, filtrare los registros, segun el tipo de cliente (tipo_cliente)

## K) Calcular la duración de cada viaje:
Crearé una nueva columna para calcular la duracion de cada viaje:

```r
tabla_general_9 <- tabla_general_8 %>%
    mutate(duracion_viaje = (fecha_hora_fin - fecha_hora_inicio))
```

Previsualizó los primeros 6 registros, para verificar que  la duracion se haya calculado correctamente

```r
head(tabla_general_9)
```

Verifico si hay registros donde la duracion del viaje sea igual o menor a 0:

```r
filtrados <- filter(tabla_general_9, duracion_viaje <= 0 )

head(filtrados)

str(filtrados)
```

Hay un total de 227 registros con una duracion de 0 segundos o que tienen duracion negativa, lo cual no es logico y deben ser descartados. Por lo que proceso a filtrar los datos de la tabla_general_9:

```r
tabla_general_10 <- filter(tabla_general_9, duracion_viaje > 0 )

head(tabla_general_10)

str(tabla_general_10)
```

Verificamos que de un total de 1.558.087, quedaron 1.557.860, descontando los 227 registros que canalizamos anteriormente

## L) Estandarizar el formato de la columna duracion_viaje, actualmente esta en segundos y lo pasare a formato Hora, minutos y segundos (HH:MM:SS):

1. Convertí los datos de segundos a periodo (H:M:S):

```r
tabla_general_10 <- tabla_general_10 %>%
  mutate(duracion_viaje_hms = seconds_to_period(duracion_viaje))

head(tabla_general_10)
```

2. Cambie el formato a HH:MM:SS:

```r
tabla_general_10 <- tabla_general_10 %>%
  mutate(duracion_viaje_formateado = sprintf("%02d:%02d:%02d",
                                            hour(duracion_viaje_hms),
                                            minute(duracion_viaje_hms),
                                            second(duracion_viaje_hms)
                                            )
        ) 

head(tabla_general_10)
```
